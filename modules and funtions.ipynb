```python
class_info = {
    "class_name": "CSE(AI)",
    "Semester": 3,
    "Subject": "SAC LAB"
}

students = [
    {"name": "Reda Naz", "roll-no": 85},
    {"name": "Grecia", "roll-no": 90},
    {"name": "Amulya", "roll-no": 103}
]

def display_class_info():
    print("Class Information:")
    for key, value in class_info.items():
        print(f"{key.capitalize()}: {value}")

def display_students():
    print("\nStudents list:")
    for s in students:
        print(f"Roll No: {s['roll-no']} - Name: {s['name']}")

# Main execution
print("College Information System\n")
display_class_info()
display_students()
```
