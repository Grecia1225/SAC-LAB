```python
import numpy as np

# Create a numpy array
arr = np.array([10, 20, 30, 40, 50])

# Perform statistical operations
mean_val = np.mean(arr)
max_val = np.max(arr)
min_val = np.min(arr)
std_dev = np.std(arr)

# Print results
print("Array:", arr)
print("Mean:", mean_val)
print("Standard Deviation:", std_dev)
print("Maximum:", max_val)
print("Minimum:", min_val)
```
