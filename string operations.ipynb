```python
greeting = "Hello"

def string_operations():
    name = "Alice"
    
    full_greeting = greeting + ", " + name + "!"
    print(full_greeting)
    
    substring1 = full_greeting[0:5]
    print("Substring 1", substring1)
    
    char_at_6 = full_greeting[6]
    print(f"Character at index 6 is {char_at_6}")

string_operations()
```
